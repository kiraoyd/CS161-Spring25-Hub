# Getting Data Into the List: appending to the end

```python
#Python has a function to add to the very end of a list

names = ["Kira", "Tommy", "Crow"]


print(f"The list starting out is: {names}")

names.append("Fardin")
print(f"The list now is: {names}")

names.append("Elisha")
print(f"The list now is: {names}")

names.append("Justin")
print(f"The list now is: {names}")

```

# Appending in a Loop

```python
# Appending items to list using a loop

names = [] #empty list

#create an input loop, that appends every name entered to the program to "names"
person = input("Enter a name, type 'quit' to stop: ")
while person != "quit":
    #add the person to my list called 'names'
    names.append(person)

    person = input("Enter a name, type quit to stop: ")

print(names)
    
```

# Accessing and Updating Individual Elements in a List

```python
names = ["Kira", "Tommy", "Crow", "Stephanie", "Mike", "Atul"]
print(names)

#access the first name, print it
#The FIRST thing in the list is always at index 0
print(f"The first name in the list is: {names[0]}")

#update the first name in the list to Ironman
#The FIRST thing in the list is always at index 0
names[0] = "Ironman"
print(f"The first name in the list is now: {names[0]}")

#Find out how many things are in our list, the size of the list
size = len(names)
print(f"There are {size} people in my names list")

#The LAST thing in a list is always at size -1
last_index = size - 1

#Change the last person in the list to "Batman"
names[last_index] = "Batman"
print(f"The list of names is now: {names}")

```
