## Simple Examples of Sentiment Analysis with TextBlob

pexels-laura-musikanski-821773-6019019.jpg

TextBlob is a Python library for processing textual data. It provides a simple API
for common natural language processing (NLP) tasks, including sentiment analysis.

Each `TextBlob` object has `polarity` and `subjectivity` properties:
- **Polarity:** A float within the range `[-1.0, 1.0]` where -1.0 is very negative and 1.0 is very positive.
- **Subjectivity:** A float within the range `[0.0, 1.0]` where 0.0 is very objective and 1.0 is very subjective.

````python
## Begin Example

# Let's analyze a positive sentence
text1 = "This movie was absolutely fantastic and I loved every minute of it!"
blob1 = TextBlob(text1)

print(f"Text 1: '{text1}'")
print(f"Sentiment for Text 1: {blob1.sentiment}")
print(f"  Polarity (how positive/negative): {blob1.sentiment.polarity:.2f}")
print(f"  Subjectivity (how opinionated): {blob1.sentiment.subjectivity:.2f}")

# Let's analyze a negative sentence
text2 = "The food was terrible and the service was extremely slow."
blob2 = TextBlob(text2)

print(f"\nText 2: '{text2}'")
print(f"Sentiment for Text 2: {blob2.sentiment}")
print(f"  Polarity: {blob2.sentiment.polarity:.2f}")
print(f"  Subjectivity: {blob2.sentiment.subjectivity:.2f}")

# Let's analyze a neutral/objective sentence
text3 = "The capital of France is Paris."
blob3 = TextBlob(text3)

print(f"\nText 3: '{text3}'")
print(f"Sentiment for Text 3: {blob3.sentiment}")
print(f"  Polarity: {blob3.sentiment.polarity:.2f}")
print(f"  Subjectivity: {blob3.sentiment.subjectivity:.2f}")
## End Example
````
